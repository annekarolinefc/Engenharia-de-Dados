# Scatter Plots

Gr√°ficos de linhas ou de pontinhos.

```python
import plotly.graph_objects as go
import numpy as np
t = np.linspace(0, 10, 100) #vetor que cria 100 pontos no intervalo de 0 a 10
y = np.sin(t) # seno de t
```

```python
t
```

```python
y
```

```python
fig = go.Figure(data=go.Scatter(x=t, y=y, mode="markers")) #pontinhos
fig.show()
```

```python
fig = go.Figure(data=go.Scatter(x=t, y=y, mode="lines")) #linha
fig.show()
```

```python
N =100
randon_x = np.linspace(0, 1, N)

#Criando 3 series
randon_y0 = randon_x + 5
randon_y1 = randon_x
randon_y2 = randon_x - 5
```

```python
fig=go.Figure()
fig.show()
```

```python
fig = go.Figure()
fig.add_traces(go.Scatter(x=randon_x, y=randon_y0, mode="markers", name="markers"))
fig.add_traces(go.Scatter(x=randon_x, y=randon_y1, mode="markers", name="markers"))
fig.add_traces(go.Scatter(x=randon_x, y=randon_y2, mode="markers", name="markers"))
```

```python
N =100
x = np.linspace(0, 1, N)

#Criando 3 series
randon_y0 = np.random.randn(N) + 5
randon_y1 = np.random.randn(N)
randon_y2 = np.random.randn(N) - 5

fig=go.Figure()

fig = go.Figure()
fig.add_traces(go.Scatter(x=randon_x, y=randon_y0, mode="markers", name="markers"))
fig.add_traces(go.Scatter(x=randon_x, y=randon_y1, mode="lines+markers", name="lines+markers"))
fig.add_traces(go.Scatter(x=randon_x, y=randon_y2, mode="lines", name="lines"))
```

```python
import plotly.graph_objects as go 

fig = go.Figure(data = go.Scatter(
    x = [1, 2, 3, 4],
    y = [10, 11, 12, 13],
    mode = "markers",
    #GRAFICO DE BOLHA
    marker=dict(
        size = [40, 60, 80, 100],
        color = [0, 1, 2, 3]
    )
))
fig.show()
```

```python
import plotly.graph_objects as go 

fig = go.Figure(data = go.Scatter(
    x = [1, 2, 3, 4],
    y = [10, 11, 12, 13],
    mode = "markers",
    #GRAFICO DE BOLHA
    marker=dict(
        size = [40, 60, 80, 100]),       
        hovertemplate= "R$ %{y} - %{marker.size}"
))
fig.show()
```
